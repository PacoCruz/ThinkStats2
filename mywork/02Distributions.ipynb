## Histograms

```python
%matplotlib inline
```

```python
# Compute frequencies with a histogram
s = 'mississippi'
hist = {}
for x in s:
    hist[x] = hist.get(x,0) + 1 # get(x,0) default value if not exist

# Can be done with defaultdict
from collections import defaultdict
d = defaultdict(int)
for k in s:
    d[k] += 1

print(hist.items())
print(d.items())
```

```python
import thinkstats2
hist = thinkstats2.Hist([1,2,2,3,5])
hist
```

```python
hist.Freq(2)
```

```python
hist.Items()
```

```python
for val, freq in hist.Items():
    print(val, freq)
```

### Plotting Histograms

```python
import thinkplot
thinkplot.Hist(hist)
thinkplot.Show(xlabel='value',ylabel='frequency')
```

```python
import first
import nsfg
```

```python
preg = nsfg.ReadFemPreg()
live = preg[preg.outcome == 1]
```

```python
hist = thinkstats2.Hist(live.birthwgt_lb, label='birthwgt_lb')
thinkplot.Hist(hist)
thinkplot.Show(xlabel='pounds', ylabel='frequency')
```

```python
hist = thinkstats2.Hist(live.birthwgt_oz, label='birthwgt_lb')
thinkplot.Hist(hist)
thinkplot.Show(xlabel='oz', ylabel='frequency')
```

```python
hist = thinkstats2.Hist(live.agepreg.round(), label='birthwgt_lb')
thinkplot.Hist(hist)
thinkplot.Show(xlabel='years', ylabel='frequency')
```

```python
hist = thinkstats2.Hist(live.prglngth, label='birthwgt_lb')
thinkplot.Hist(hist)
thinkplot.Show(xlabel='weeks', ylabel='frequency')
```

### Outliers

```python
for weeks, freq in hist.Smallest(10):
    print(weeks, freq)
```
