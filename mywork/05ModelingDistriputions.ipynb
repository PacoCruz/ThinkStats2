```python
%matplotlib inline

import numpy as np

import nsfg
import first
import thinkstats2
import thinkplot
import analytic
```

```python
df = analytic.ReadBabyBoom()
diffs = df.minutes.diff()
cdf = thinkstats2.Cdf(diffs, label='actual')

thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='minutes', ylabel='CDF')
```

## The exponential distribution

The CDF of the exponential distribution is simple:

$$CDF(x)=1-e^{-\lambda x}$$

This distribution comes up when we look at a series of events and we measure the times between events, called **interarrival times**.

To confirm that the above CDF plot can be modeled with an exponential distribution, we can use the **complementary CDF**, which is $1-CDF(x)$ and plot on a log-y scale.

```python
thinkplot.Cdf(cdf, complement=True)
thinkplot.Show(xlabel='minutes', ylabel='CDG',
               yscale='log')
```
